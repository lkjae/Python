```python
#심화1
n = 5000
num = []
for i in range(n+1):
    result = True
    if(i<2): result = False
    for j in range(2,i):
        if(i%j==0):
            result = False
    if result:
        num.append(i)
print(num)
```

```python
# 제곱근
# x의 제곱근 xx.xxxx
# xx.xxxx >= 양의정수
# 나누어 떨어지면 소수가 아니다.
import math

primenumber = []
for i in range(2,5001):
    divide = 0
    for j in range(2,int(math.sqrt(i)) + 1):
        if(i%j == 0):
            divide += 1
    if divide == 0:
        primenumber.append(i)
print(primenumber)
```

```python
#심화2
def prime():
    n = int(input('2 이상의 정수 입력: '))
    num = []
    for i in range(n+1):
        result = True
        if(i<2): result = False
        for j in range(2,i):
            if(i%j==0):
                result = False
        if result:
            num.append(i)
    return num
```

```python
s = "Big Data Analysisi"
def alpha_max(s):
    alphabet = {}
    for i in s:
        if i not in alphabet:
            alphabet[i] = 1
        else:
            alphabet[i] += 1

    alphabet_list = []
    for k,v in alphabet.items():
        if v == max(alphabet.values()):
            alphabet_list.append(k)
            
    return alphabet_list
```
