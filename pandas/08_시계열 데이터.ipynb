```python
banks = pd.read_csv('data/banklist.csv', parse_dates=[5, 6])
```

```python
banks.info()
```

```python
banks.head()
```

```python
banks['closing_quarter'] = banks['Closing Date'].dt.quarter
banks['closing_year'] = banks['Closing Date'].dt.year
banks
```

```python
closing_year_q = banks.groupby(['closing_year', 'closing_quarter'])['City'].count()
```

```python
closing_year_q.plot()
```

```python
closing_year_q.index[:5]
```

```python
closing_year_q.values
```

```python
banks.groupby('closing_year')['City'].count().plot()
```

```python
banks['closing_month'] = banks['Closing Date'].dt.month
```

```python
banks.groupby(['closing_year', 'closing_month'])['City'].count().plot()
```
